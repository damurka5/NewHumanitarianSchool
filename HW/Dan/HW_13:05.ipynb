```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)
sin_y = np.sin(x)
cos_y = np.cos(x)

fig, (ax1, ax2) = plt.subplots(nrows=2, sharex=True)
fig.suptitle('Синус и косинус')

ax1.plot(x, sin_y, label='sin(x)')
ax1.set_ylabel('y')
ax1.legend()

ax2.plot(x, cos_y, label='cos(x)')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Создание сигнала
Fs = 1000  # Частота дискретизации
t = np.arange(0, 1, 1/Fs)  # Временная ось
f1 = 5  # Частота первой гармоники
f2 = 50  # Частота второй гармоники
signal = np.sin(2 * np.pi * f1 * t) + np.sin(2 * np.pi * f2 * t)  # Сумма двух синусоид

# Вычисление коэффициентов разложения по анализу Фурье
fft = np.fft.fft(signal)  # Прямое преобразование Фурье
freqs = np.fft.fftfreq(signal.size, 1/Fs)  # Вычисление частот

# Визуализация результатов
fig, (ax1, ax2) = plt.subplots(nrows=2, sharex=True, figsize=(8, 6))
fig.suptitle('Разложение по анализу Фурье')

ax1.plot(t, signal)
ax1.set_xlabel('Время, с')
ax1.set_ylabel('Амплитуда')
ax1.set_title('Исходный сигнал')

ax2.stem(freqs, np.abs(fft)/len(signal))
ax2.set_xlabel('Частота, Гц')
ax2.set_ylabel('Амплитуда')
ax2.set_xlim(-100, 100)
ax2.set_title('Амплитудный спектр')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Генерация данных
t = np.linspace(0, 2 * np.pi, 200)
f = np.sin(t) + np.sin(3*t)

# Разложение Фурье
n = 50
a0 = np.mean(f)
an = np.zeros(n)
bn = np.zeros(n)

for i in range(1, n+1):
    an[i-1] = (2/np.pi)*np.trapz(f*np.cos(i*t), t)
    bn[i-1] = (2/np.pi)*np.trapz(f*np.sin(i*t), t)

# Анимация разложения Фурье
fig, ax = plt.subplots()
ax.set_xlim(0, n+1)
ax.set_ylim(-2, 2)
ax.set_xlabel('n')
ax.set_ylabel('Cn')

line, = ax.plot([], [], 'bo')

def update(i):
    x = np.arange(1, i+1)
    y = np.concatenate(([a0], 2*an[:i-1], [an[i-1]], np.zeros(n-i), 2*bn[:i-1], [bn[i-1]], np.zeros(n-i)))
    line.set_data(x, y)
    return line,

ani = FuncAnimation(fig, update, frames=n+1, blit=True)

# Отображение анимации
plt.show()
```
